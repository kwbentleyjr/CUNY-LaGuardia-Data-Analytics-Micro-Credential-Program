# Task#4: Data Cleaning/ Feature Engineering

```python
# Let's drop the date
df_alexa = df_alexa.drop(['date', 'rating'],axis=1)
```

```python
df_alexa
```

```python
variation_dummies = pd.get_dummies(df_alexa['variation'], drop_first = True)
```

```python
variation_dummies
```

```python
# first let's drop the column
df_alexa.drop(['variation'], axis=1, inplace=True)
```

```python
# Now let's add the encoded column again
df_alexa = pd.concat([df_alexa, variation_dummies], axis=1)
```

```python
df_alexa
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
alexa_countvectorizer = vectorizer.fit_transform(df_alexa['verified_reviews'])
```

```python
alexa_countvectorizer.shape
```

```python
type(alexa_countvectorizer)
```

```python
print(vectorizer.get_feature_names())
```

```python
print(alexa_countvectorizer.toarray())  
```

```python
# first let's drop the column
df_alexa.drop(['verified_reviews'], axis=1, inplace=True)
reviews = pd.DataFrame(alexa_countvectorizer.toarray())
```

```python
# Now let's concatenate them together
df_alexa = pd.concat([df_alexa, reviews], axis=1)
```

```python
df_alexa
```

```python
# Let's drop the target label coloumns
X = df_alexa.drop(['feedback'],axis=1)
```

```python
y = df_alexa['feedback']
```

# Task #5: Model Building and Model Training

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

randomforest_classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
randomforest_classifier.fit(X_train, y_train)
```

# Task #6: Model Evaluation

```python
y_predict_train = randomforest_classifier.predict(X_train)
cm = confusion_matrix(y_train, y_predict_train)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_train, y_predict_train))
```

```python
y_predict = randomforest_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

# Conclusion

When exploring the Amazon review dataset, we see that 257 reviews are considered negative and 2893 reviews are categorized as positive.  Of the positive reviews an overwhelming number received the highest rating of 5.  When plotting the ratings of the different variations, we see that they all had an average rating of at least 4.

To clean the data and prepare for the sentiment analysis on the feedback column, we first dropped the ratings and date columns as they are not needed.  We then used the get_dummies function to turn the variation column into numbers.  Next, we used the CountVectorizer function to convert the text into numbers.

To build our model we split the data into two datasets.  We use 80% of the data to train and 20% of the data to test our model.  To create the first model, we used the RandomForestClassifier function with entropy algorithm. The entropy algorithm measures the randomness in the system, and we use this to split the nodes in the decision tree.  Using the predict method of RandomForestClassifer we create an array of the predicted values.

From the classification report on the first model, we see that the positive review training data had both high precision and high recall.  This means that of the total positive review predictions, 92% were correct (precision).  Also, of the total number of positive reviews in the dataset, our model predicted 99% of them.  However, our first model was far less accurate at predicting precision and recall for negative reviews.  Of all the negative review predictions, only 73% were correct (precision).  Of the total negative reviews in the dataset, our first model only predicted 14% of them. 

In our second model we dropped the variation columns and added a column for the length of the reviews.  We used the same entropy algorithm, but we increased the number of n estimators to 300 from 100. This means it will increase the number of decision trees in the forest from 100 to 300.  The second model had better precision and recall for both classes (positive, negative reviews).  However, the second model was still poor at predicting class 0 (negative reviews).

Most reviews in the dataset are positive. This may account for the poor predictions of negative reviews in both models. Adding the review length as an independent variable helped the precision of both models. However, the models are poor at predicting negative reviews when the reviews are true negative (poor recall).
